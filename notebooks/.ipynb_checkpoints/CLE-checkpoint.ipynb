```python
import scipy as sp
import glob
# fnames = glob.glob("../data/CLE/longtraj_HSC_nonec*_dim_11.mat")
fnames = ["../data/CLE/longtraj_HSC_nonec0.1_dim_11.mat",
          "../data/CLE/longtraj_HSC_nonec0.3_dim_11.mat", 
          "../data/CLE/longtraj_HSC_nonec0.5_dim_11.mat", 
          "../data/CLE/longtraj_HSC_nonec0.7_dim_11.mat", 
          "../data/CLE/longtraj_HSC_nonec0.9_dim_11.mat"
         ]
mats = [sp.io.loadmat(f) for f in fnames]
```

```python
import os
import seaborn as sb
import pandas 
import matplotlib.pyplot as plt
plt.figure(figsize = (9, 3))
for i in range(len(mats)):
    plt.subplot(2, 3, i+1)
    plt.plot(mats[i]['samples'][:, 1, :-1], alpha = 0.5)
```

```python
import numpy as np
Xs = []
ts = []
inds = np.random.randint(0, mats[0]['samples'].shape[0], 1_000)
for mat in mats:
    X = np.vstack([mat['samples'][inds[i], i, :] for i in range(len(inds))])
    t = X[:, -1]
    X = X[:, :-1]
    X = X[np.argsort(t), :]
    t = t[np.argsort(t)]
    ind = X.sum(-1) > 10
    Xs.append(X[ind, :])
    ts.append(t[ind])
```

```python
for i in range(len(Xs)):
    plt.subplot(2, 3, i+1)
    sb.heatmap(Xs[i])
```

```python
import sklearn as sk
import sklearn.decomposition
import umap
pca_op=sk.decomposition.PCA()
pca_op.fit(Xs[0])
Xs_pca = [pca_op.transform(X) for X in Xs]
Xs_umap = [umap.UMAP(n_neighbors=50, min_dist = 0.7).fit_transform(X) for X in Xs]
```

```python
for i in range(len(Xs)):
    plt.subplot(2, 3, i+1)
    sb.scatterplot(x = Xs_pca[i][:, 1], y = Xs_pca[i][:, 2], alpha = 0.1, hue = ts[i], palette = 'RdBu_r', legend = False)
    plt.xlim(-6, 10)
    plt.ylim(-7, 7)
```

```python
Ws = []
us = []
Ws_ent = []
us_ent = []
for X in Xs:
    np.save("X.npy", X)
    !julia --project=.. ../src/kernel_cmd.jl X.npy 5.0 --mode quad --out_W W_quad.npy --out_u u_quad.npy --diag_inf true
    W, u = np.load("W_quad.npy"), np.load("u_quad.npy")
    Ws.append(W)
    us.append(u)
    !julia --project=.. ../src/kernel_cmd.jl X.npy 0.05 --mode ent --out_W W_ent.npy --out_u u_ent.npy --diag_inf true
    W, u = np.load("W_ent.npy"), np.load("u_ent.npy")
    Ws_ent.append(W)
    us_ent.append(u)
```

```python
from tqdm import tqdm
Us = [np.linalg.eigh(W)[1][:, ::-1] for W in tqdm(Ws)]
Us_ent = [np.linalg.eigh(W)[1][:, ::-1] for W in tqdm(Ws_ent)]
# kNN
Ws_knn = [sk.neighbors.NearestNeighbors(n_neighbors=25).fit(X).kneighbors_graph() for X in Xs_pca]
Ws_knn = [np.asarray((A + A.T).todense())/2 for A in Ws_knn]
Us_knn = [np.linalg.eigh(W)[1][:, ::-1] for W in tqdm(Ws_knn)]
```

```python
plt.figure(figsize = (10, 2))
for i in range(5):
    plt.subplot(1, 5, i+1)
    sb.heatmap(Us[i][:, :10], vmin = -0.1, vmax = 0.1, cmap = "RdBu_r")

plt.figure(figsize = (10, 2))
for i in range(5):
    plt.subplot(1, 5, i+1)
    sb.heatmap(Us_ent[i][:, :10], vmin = -0.1, vmax = 0.1, cmap = "RdBu_r")
    
plt.figure(figsize = (10, 2))
for i in range(5):
    plt.subplot(1, 5, i+1)
    sb.heatmap(Us_knn[i][:, :10], vmin = -0.1, vmax = 0.1, cmap = "RdBu_r")
```

```python
def choose_vec(x, t):
    # return x if np.dot(t, x) > 0 else -x
    return x if x[t < np.quantile(t, 0.5)].mean() > 0 else -x
```
