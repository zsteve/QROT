```python
import os
eps_ent = np.sort([float(x) for x in pd.Series([os.path.splitext(x)[0] for x in glob.glob("*ent_eps*.csv")]).str.split('_').str[-1].unique()])
eps_l2 = np.sort([float(x) for x in pd.Series([os.path.splitext(x)[0] for x in glob.glob("*l2_eps*.csv")]).str.split('_').str[-1].unique()])
```

```python
import glob
def symm_norm(x):
    return x * (np.sum(x, -1) ** (-0.5))[None, :] * (np.sum(x, -1) ** (-0.5))[:, None]
K = symm_norm(pd.read_csv("K.csv").values)


ranks = np.sort(pd.Series(glob.glob("K_nys_*_proj_l2_eps_0.5.csv")).str.split('_').str[2].astype(int).values)

K_proj_l2 = [pd.read_csv(f"K_proj_l2_eps_{x}.csv").values for x in tqdm(eps_l2)]
K_proj_ent = [pd.read_csv(f"K_proj_ent_eps_{x}.csv").values for x in tqdm(eps_ent)]
K_nys = [symm_norm(pd.read_csv(f"K_nys_{rank}.csv").values) for rank in tqdm(ranks)]
K_nys_proj_l2 = [[pd.read_csv(f"K_nys_{rank}_proj_l2_eps_{x}.csv").values for rank in ranks] for x in tqdm(eps_l2)]
K_nys_proj_ent = [[pd.read_csv(f"K_nys_{rank}_proj_ent_eps_{x}.csv").values for rank in ranks] for x in tqdm(eps_ent)]
```

```python
from tqdm import tqdm
import scipy as sp
d = 25
def get_eigs(x, d):
     return np.linalg.eigh(np.eye(x.shape[0]) - x)[1][:, range(d)]

thetas = []
U_ref = get_eigs(K, d)
thetas = [sp.linalg.subspace_angles(U_ref, get_eigs(x, d)).mean() for x in tqdm(K_nys)]

U_ref_ent = [get_eigs(x, d) for x in K_proj_ent]
thetas_ent = [[sp.linalg.subspace_angles(U, get_eigs(x, d)).mean() for x in Ks] for (U, Ks) in tqdm(zip(U_ref_ent, K_nys_proj_ent))]

U_ref_l2 = [get_eigs(x, d) for x in K_proj_l2]
thetas_l2 = [[sp.linalg.subspace_angles(U, get_eigs(x, d)).mean() for x in Ks] for (U, Ks) in tqdm(zip(U_ref_l2, K_nys_proj_l2))]
```

```python
import math
plt.figure(figsize = (10, 3))
plt.subplot(1, 3, 1)
_df_kernel = pd.DataFrame(thetas, index = ranks, columns = ['theta']).reset_index(names = 'rank')
sb.lineplot(data = _df_kernel, x = 'rank', y = 'theta', color = 'red')
plt.title("Symmetric")
plt.ylim(0.05, math.pi/2); plt.yscale('log')
plt.subplot(1, 3, 2)
_df = pd.melt(pd.DataFrame(thetas_ent, index = eps_ent, columns = ranks).reset_index(names = 'eps'), id_vars = 'eps', var_name = 'rank', value_name = 'theta')
sb.scatterplot(data = _df, x = 'rank', y = 'theta', hue = 'eps')
sb.lineplot(data = _df_kernel, x = 'rank', y = 'theta', linestyle = 'dashed', color = 'red')
plt.ylim(0.05, math.pi/2); plt.yscale('log')
plt.title("Entropy")
plt.subplot(1, 3, 3)
_df = pd.melt(pd.DataFrame(thetas_l2, index = eps_l2, columns = ranks).reset_index(names = 'eps'), id_vars = 'eps', var_name = 'rank', value_name = 'theta')
sb.scatterplot(data = _df, x = 'rank', y = 'theta', hue = 'eps')
sb.lineplot(data = _df_kernel, x = 'rank', y = 'theta', linestyle = 'dashed', color = 'red')
plt.ylim(0.05, math.pi/2); plt.yscale('log')
plt.title("L2")
```

```python
# calculate perplexity
def perplexity(M):
    return np.exp(-(sp.special.xlogy(M, M)).sum(-1))
perp_ent = [perplexity(x).mean() for x in K_proj_ent]
perp_l2 = [perplexity(x).mean() for x in K_proj_l2]
```

```python
rank_idx = 4
xs = np.array([perplexity(y[rank_idx]).mean() for y in K_nys_proj_l2])
ys = [x[rank_idx] for x in thetas_l2]
plt.plot(xs, ys, marker = '^', label = "L2")
xs = np.array([perplexity(y[rank_idx]).mean() for y in K_nys_proj_ent])
ys = [x[rank_idx] for x in thetas_ent]
plt.plot(xs, ys, marker = 'o', label = "Entropic")
plt.xlabel("Perplexity")
plt.ylabel("Subspace angle")
plt.xscale("log");
plt.yscale("log");
plt.title(f"Nystrom rank {ranks[rank_idx]}")
plt.legend()
```

```python
import networkx as nx
import sklearn as sk

def get_partitions(_g):
    partitions = nx.community.louvain_communities(_g, resolution = 1)
    partition_index = np.zeros(K.shape[0], dtype = np.int64)
    for (i, p) in enumerate(partitions):
        partition_index[list(p)] = i
    return partitions, partition_index
```
