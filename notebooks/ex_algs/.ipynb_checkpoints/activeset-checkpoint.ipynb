```julia
using Pkg
Pkg.activate(".")
using OptimalTransport
using LinearAlgebra
using SparseArrays
using Distances
using StatsBase
using NNlib
using BenchmarkTools
using NearestNeighbors
using RandomMatrix
using ParallelNeighbors
using ProgressMeter

include("../../src/util.jl")
```

```julia
using DataFrames
using CSV
using Glob
using StatsPlots

dfs_time = []
dfs_bytes = []

d = 250

for N in [1000, 5000, 10_000, 25_000,]
    files = Glob.glob("output/runtimes_*_N_$(N)_d_$(d)_*.csv")
    _df = [CSV.read(x, DataFrame)[2:end, :] for x in files]
    push!(dfs_time, DataFrame(:Dense=> [mean(x.dense_time) for x in _df], :ActiveSet => [mean(x.sparse_time) for x in _df], :N => N))
    push!(dfs_bytes, DataFrame(:Dense=> [mean(x.dense_bytes) for x in _df], :ActiveSet => [mean(x.sparse_bytes) for x in _df], :N => N))
end
```

```julia
df_time = stack(vcat(dfs_time...));
df_bytes = stack(vcat(dfs_bytes...));
df_time = df_time[.!ismissing.(df_time.value), :]
df_bytes = df_bytes[.!ismissing.(df_bytes.value), :];
```

```julia
plt1 = @df df_time groupedboxplot(:variable, :value, group = :N; ylim = (-10, 650), palette = :Reds_5, ylabel = "Time (s)", legend = :outertopright)
plt2 = @df df_bytes groupedboxplot(:variable, :value, group = :N; ylim = (-0.01e12, 0.9e12), palette = :Reds_5, ylabel = "Memory (bytes)", legend = :outertopright);
plt=plot(plt1, plt2; size = (750, 400), margin = 3*Plots.mm)
savefig("../figures/activeset_benchmark.pdf")
plt
```
