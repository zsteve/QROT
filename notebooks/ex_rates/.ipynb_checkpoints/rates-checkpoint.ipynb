```julia
# d-sphere
function get_plot(N)
    # s = range(-, ; length = N)
    # X = mapreduce(x -> [cos(x), sin(x)], hcat, s)
    X = randn(d, N)
    X ./= reshape(map(norm, eachcol(X)), 1, :)
    @info size(X)
    C = pairwise(SqEuclidean(), X)
    # C /= mean(C)
    vals = 10 .^range(6, -2; length = 100)
     = fill(1/size(X, 2), size(X, 2))
    function dualpotentials()
        # @info 
        solver = OptimalTransport.build_solver(, C, , OptimalTransport.SymmetricQuadraticOTNewton( = 1e-5); maxiter = 100)
        OptimalTransport.solve!(solver)
        solver.cache.u
    end
    uvals = map(x -> dualpotentials(x), vals)
    umeans = map(mean, uvals)
    # df1 = DataFrame(logu = log10.(umeans[vals .< 0.05]), logeps = log10.(vals[vals .< 0.05]))
    # fit1 = lm(@formula(logu ~ logeps), df1)
    df2 = DataFrame(logu = log10.(umeans[vals .> 25]), logeps = log10.(vals[vals .> 25]))
    fit2 = lm(@formula(logu ~ logeps), df2)
    plt=scatter(log10.(vals), log10.(umeans); markersize = 1, alpha = 0.25, label = "data", legend = :topleft, title = "N = $N")
    # plot!(plt, df1.logeps, predict(fit1); label = " = $(coef(fit1)[2])")
    plot!(plt, df2.logeps, predict(fit2); label = " = $(round(coef(fit2)[2]; digits = 3))")
    # hline!(plt, [log10.(Cthresh), ]; label = "u_thresh")
    plt
end
```

```julia
x0 = [0, 1, 0.5]
d = 2
f1(x) = dot([3, 5, 7], x.^2 /2)
r, R = 0.5, 1
```

```julia
function sample(f; N = 250, C0 = 0.05,  = 1.5)
    X = vcat(x0', convert(Array, torusUnif(N, r, R)))'
    y = map(f, eachcol(X))
    C = pairwise(SqEuclidean(), X); C[diagind(C)] .= Inf
     = fill(1/size(X, 2), size(X, 2));
     = C0*N^ # pick exponent 3/2
    K_eps_N = ^(2/(d+2)) * N^(-4/(d+2))
    # solve QOT
    solver = OptimalTransport.build_solver(, C, , OptimalTransport.SymmetricQuadraticOTNewton( = 1e-5); maxiter = 25)
    OptimalTransport.solve!(solver)
    W = OptimalTransport.plan(solver) * size(X, 2)
    L = I - W
    -2*(L * y)[1] / K_eps_N
end
```

```julia
using ProgressMeter
using StatsPlots
using LaTeXStrings
M = 25
Ns = [100, 250, 500, 1_000, 2_500, 5_000, 10_000]
C0 = 1.0
# Ns = [100, 250, 500, 1000, 2500, 5000, 10_000]
plts_all = []
for  in [2.0, 1.75, 1.5, 1.25]
    reval("set.seed(0)")
    L_samples = [@showprogress [sample(f1; N = k,  = , C0 = C0) for _ in 1:M] for k in Ns]
    # what is the effective epsilon?
    eps_eff = [C0*k^ / k^2 for k in Ns]
    plt=boxplot(hcat(L_samples...), xticks = (1:length(Ns), Ns), color = :lightgrey, xlabel = L"N", ylabel = L"K_{, N}^{-1} (\Delta^{OT} y)(x_0)", legend = nothing, title = latexstring(" = $()"), linecolor = :black)
    push!(plts_all, plt)
    savefig("qot_torus_alpha_$().pdf")
end
```
